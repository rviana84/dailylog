```python
import pandas as pd
```

```python
# Leitura do arquivo original extraído do BigQuery
df = pd.read_csv('original_carol.csv')
```

```python
# Informações do arquivo
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# Separação em duas categorias: ampla e não-ampla
ampla = df[df['modalidade'] == 'Ampla concorrência.']
cota = df[df['modalidade'] != 'Ampla concorrência.']
```

```python
# Verificação
ampla.info()
```

```python
cota.info()
```

```python
# Criação da coluna no df 'ampla' com a menor nota de cada curso
ampla['nota_minima'] = ampla.groupby(['id_curso'])['nota'].transform('min')
```

```python
# Checagem
ampla.tail(20)
```

```python
# Transformação em lista
lista = list(zip(ampla['id_curso'], ampla['nota_minima']))
```

```python
# Conversão em df
ampla_minima = pd.DataFrame(lista, columns=['id_curso', 'meta'])
```

```python
# Remoção de dados duplicados
ampla_minima.drop_duplicates(inplace=True)
```

```python
# Fusão de dfs 'ampla_minima' e 'cota'
final = cota.merge(ampla_minima, on='id_curso', how='inner')
```

```python
# Recorte de notas maiores (em não-ampla) que a mínima (de ampla) de cada curso
final = final[final['nota'] > final['meta']]
```

```python
# Verificação
final.info()
```

```python
final['id_curso'].nunique()
```

```python
final['modalidade'].value_counts()
```

```python
# Gravação em csv
final.to_csv('cotistas_que_passariam_na_ampla.csv', encoding='utf-8', index=False)
```
