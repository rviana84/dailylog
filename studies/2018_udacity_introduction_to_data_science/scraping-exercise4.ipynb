```python
import glob
import pandas as pd

df_list = []
for i in glob.glob('data/ebert_reviews/*.txt'):
    with open(i, encoding='utf-8') as f:
        title = f.readline()[:-1]
        url = f.readline()[:-1]
        text = f.read()
        df_list.append({'title': title, 'url': url, 'text': text})
df = pd.DataFrame(df_list, columns=['title', 'url', 'text'])
```
