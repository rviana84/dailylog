```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as r

url = 'https://www.al.sp.gov.br/repositorio/folha-de-pagamento/folha-2018-10.html'
data = r.get(url).content
soup = BeautifulSoup(data, 'html.parser')
```

```python
lista = []
for tr in soup.find_all('tr')[1:]:
    col = 0
    for i in tr.find_all('td'):
        td_limpo = i.get_text().strip()
        if col == 0:
            nome = td_limpo
        if col == 1:
            bruto = td_limpo.replace('.', '').replace(',', '.')
        if col == 2:
            liquido = td_limpo.replace('.', '').replace(',', '.')
            lista.append({'nome': nome, 
                          'vlr_bruto': float(bruto), 
                          'vlr_liquido': float(liquido)})
        col += 1
df = pd.DataFrame(lista, columns = ['nome', 'vlr_bruto', 'vlr_liquido'])
```
